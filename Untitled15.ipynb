```python
import pandas as pd
import numpy as np
from tensorflow. keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_auc_score, f1_score
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
data = pd.read_csv('/content/drive/MyDrive/wine_quality.csv')
```

```python
data
```

```python
features = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',	'chlorides',	'free sulfur dioxide',	'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol']
target = 'quality'
```

```python
print(data.columns)
```

```python
# Convert categorical features to numerical
data['quality'] = data['quality'].map({'low': 0, 'high': 1})
```

```python
data.isnull().sum()
```

```python
data.dtypes
```

```python
X = data[features].values
y = data[target].values
```

```python
#정규화
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
#훈련 데이터셋과 테스트 데이터셋으로 분류
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
#심층신경망 모델링
model = Sequential([
      Dense(64, input_shape=(X_train.shape[1],), activation='relu'),
      Dense(32, activation='relu'),
      Dense(1, activation='sigmoid')
])
```

```python
model.summary
```

```python
#학습률 설정
model.compile(optimizer=Adam(learning_rate=0.001),
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
#학습 수행(훈련 반복수와, batchsize(한번에 몇 개의 데이터를 수행할건지))
model.fit(X_train, y_train, epochs=10, batch_size=32)
```

```python
#예측확률
y_test_pred_prob = model.predict(X_test)
```

```python
#예측성능평가
y_pred = (y_test_pred_prob > 0.5).astype('float32')
y_pred
```

```python
accuracy = accuracy_score(y_test, y_pred)
accuracy
print(accuracy)
```
